### 1.1 Input & Output

#### Screen input

Read a string from the screen (optionally, with a prompt):

````python
x = input("Enter a number:")
print("You just enter",x)
print(type(x))
````

````python
x = float(input("Enter a number:"))
print("You just enter",x)
print(type(x))
````

#### Screen output
- ```print```: the most popular way to print to screen

````python
# print two objects of different types
five = 5
print("My number is ", five)       
````

````python
# print a concatenated string with +
print("My number is " + str(five))  
````

````python
# print a formatted string using format function and {}
print("My numbers are {} and {}".format(five,1))  
````

````python
# print a formatted string
print("The sum of {} and {} is {}".format(five,1,five+1))  
````

````python
# write in different way
str2 = f"the sum of {five} and {1} is {five+1}"
print(str2)
````

#### Read and write from/to file
```open``` function returns a file object

````python
# Open a new file to write in
f = open('textfile.txt','w')
for n in range(five):
    f.write(str(n)+'\n')
f.close()   
````

````python
# Open a file to read in
f = open('textfile.txt','r')
for line in f:    
    print(line)    
f.close()    
````
